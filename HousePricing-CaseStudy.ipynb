```python
data = pd.read_csv('house_prices.csv')
house_prize_df = pd.DataFrame(data)
house_prize_df
```

# **DATA CLEANSING**

```python
#to check the missing values from each column
missing_values = house_prize_df.isnull().sum()
missing_values 
```

```python
Missings_data = house_prize_df.dropna(inplace=False)
Missings_row= pd.DataFrame(Missings_data)
Missings_row #to drop the rows which are missing data in it and not that important
```

```python
#To delete the Unnecessary columns
colums_to_drop = ['Dimensions', 'Plot Area', 'Index']
house_prize_df.drop(columns = colums_to_drop, inplace= True)    
```

```python
#To add some data in the description column
default_description = "No description available."
house_prize_df['Description'].fillna(default_description, inplace=True)
```

```python
#to fill the missing values in the price column putting mean of the price
price_mean = house_prize_df['Price'].mean()
house_prize_df['Price'].fillna(price_mean, inplace=True) 
```

```python
#To add some data in the Transaction column
default_transaction = "No Data available." 
house_prize_df['Transaction'].fillna(default_transaction, inplace=True)
```

```python
#To add some data in the Status column
default_status = 'No Data available'
house_prize_df.Status.fillna(default_status,inplace=True)
```

```python
#To add some data in the Floor column
default_floor = 'No Data available'
house_prize_df.Floor.fillna(default_floor, inplace= True)
```

```python
#To add some data in the Furnishing column
default_furnishing = 'No Data available'
house_prize_df.Furnishing.fillna(default_furnishing, inplace=True)
```

```python
#To add some data in the Facing column
default_facing =  'No Data available'
house_prize_df.facing.fillna(default_facing , inplace= True)
```

```python
#To add some data in the overlooking column
default_overlooking = 'No Data available'
house_prize_df.overlooking.fillna(default_facing , inplace=True)
```

```python
#To add some data in the society column
default_Society = 'No Data available'
house_prize_df.Society.fillna(default_Society, inplace=True)
```

```python
#To add some data in the Car parking column
default_CarParking = 'No Data available'
house_prize_df['Car Parking'].fillna(default_Society, inplace=True)
```

```python
#To add some data in the Owenership column
default_Ownership = 'No Data available'
house_prize_df.Ownership.fillna(default_Society, inplace=True)
```

*Now we want to covert the data-types of some columns to the correct data-types, but some implications come like if you wan to convert 'object' to 'int64' and some values like 'NAN' it cannot be directly convert you have to remove the 'NAN' and add '0'  or  if we have unit like 'sqrt, area' we have remove it. We will use fillna to do it*

```python
#In this we area just removing the sqrt
house_prize_df['Carpet Area'] = pd.to_numeric(house_prize_df['Carpet Area'].str.replace(r'[^\d]', '', regex=True), errors='coerce')
```

```python
#To fill the missing values in the Carpet Areacolumn putting mean of the Carpet Area
carpet_mean = house_prize_df['Carpet Area'].mean()
house_prize_df['Carpet Area'].fillna(carpet_mean, inplace=True)   
```

```python
#In this we filled the missing values to the columns
default_value = 0
house_prize_df[['Bathroom', 'Balcony']] = house_prize_df[['Bathroom', 'Balcony']].fillna(default_value)
```

```python
#In this we converted the data-type
columns_to_convert = ['Bathroom', 'Balcony']
house_prize_df[columns_to_convert] = house_prize_df[columns_to_convert].apply(pd.to_numeric, errors='coerce', downcast='integer')
```

```python
#To fill the missing values in the Bathroom column putting mean of the Bathroom
bathroam_mean = house_prize_df.Bathroom.mean()
house_prize_df.Bathroom.fillna(bathroam_mean, inplace= True)
```

```python
#to fill the missing values in the balcony column putting mean of the Balcony
Balcony_mean = house_prize_df.Balcony.mean()
house_prize_df.Balcony.fillna(Balcony_mean, inplace= True) 
```

```python
#In this we area just removing the sqrt
house_prize_df['Super Area'] = pd.to_numeric(house_prize_df['Super Area'].str.replace(r'[^\d]', '', regex=True), errors='coerce')
```

```python
#to fill the missing values in the balcony column putting mean of the Super area
super_area_mean =  house_prize_df['Super Area'].mean()
house_prize_df['Super Area'].fillna(super_area_mean, inplace=True)
```

```python
import pandas as pd
import re

# Assuming you have loaded your dataset into the DataFrame `house_prize_df`

# Function to extract the BHK value from the 'Title' string
def extract_bhk(title_str):
    bhk_match = re.search(r'\b(\d+)\s*BHK\b', title_str)
    if bhk_match:
        return int(bhk_match.group(1))
    else:
        return None

# Apply the function to create the 'BHK' column
house_prize_df['BHK'] = house_prize_df['Title'].apply(extract_bhk)
```

```python
default_value = 0
house_prize_df['BHK'].fillna(default_value, inplace=True)
house_prize_df['BHK'] = house_prize_df['BHK'].astype('int64')
```
